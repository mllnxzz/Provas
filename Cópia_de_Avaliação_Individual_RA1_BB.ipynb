# Quest√£o 1: 

Escreva uma fun√ß√£o em Python que compute o valor da fun√ß√£o abaixo, emitindo uma
mensagem de erro caso a fun√ß√£o receba um valor n√£o permitido. Al√©m disso, a fun√ß√£o dever√°
plotar um gr√°fico da fun√ß√£o no intervalo entre ¬±10, destacando neste gr√°fico os pontos
relacionados ao valor de passado para a fun√ß√£o.
ùëì(ùë•) = 1 / ‚àö2 ‚àí ùë•

```python
import math
import matplotlib.pyplot as plt
import numpy as np


def f(x):
    if x >= 2:
        raise Exception("Valor inv√°lido.")
    else:
        return 1 / math.sqrt(2 - x)


vetX = np.linspace(-10, 1, 1000)
vetY = [f(i) for i in vetX]

i = float(input("Digite um valor -> "))


plt.plot(vetX, vetY)
plt.scatter(i, f(i), color="red")
plt.show()
```

# Quest√£o 2
Utilizando a linguagem Python, e somente as bibliotecas Matplotlib e Numpy, crie uma
fun√ß√£o para plotar uma raquete de t√™nis de mesa. No seu gr√°fico a raquete deve ser plotada
de forma que seja poss√≠vel observar tanto o cabo quanto a √°rea de rebatimento.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches


def plot_table_tennis_racket():
    fig, ax = plt.subplots()

    # Cabo da raquete
    handle = patches.Rectangle((1.77, -1), 0.5, 2, facecolor='black')
    ax.add_patch(handle)

    # √Årea de rebatimento (c√≠rculo)
    rebound_area = patches.Circle(
        (2, 2.5), 2, facecolor='red', edgecolor='black', linewidth=1)
    ax.add_patch(rebound_area)

    # Configura√ß√µes do gr√°fico
    ax.set_xlim(-1, 5)
    ax.set_ylim(-1, 5)
    ax.set_aspect('equal', adjustable='box')
    plt.show()


# Chamar a fun√ß√£o
plot_table_tennis_racket()
```

# Quest√£o 3
Utilizando a linguagem Python, nenhuma biblioteca extra e List Comprehension crie uma fun√ß√£o que devolva o resultado da f√≥rmula a seguir para ùëõ elementos.

Captura de tela 2023-04-24 205830.png

```python
def f(n):
    return 1 / (1 + (2 ** n))

def somatorio(n):
    soma = 0
    for i in range(1, n+1):
        soma += f(i)
    return soma

def resposta(n):
    return 4 * (1 - somatorio(n))

i = input("Digite um valor: ")

print(resposta(i))
```

ùëì(ùë•) = 3ùë•2 ‚àí 1 / ‚àöùë•

```python
def limite(x):
    n = 3 * (x ** 2) - 1
    d = x ** (1/2) + x

    if (x == 0):
        return

    return (n/d)


i = input("Digite um valor: ")
print(limite(i))
```
