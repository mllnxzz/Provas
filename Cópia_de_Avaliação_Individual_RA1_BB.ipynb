# Questão 1: 

Escreva uma função em Python que compute o valor da função abaixo, emitindo uma
mensagem de erro caso a função receba um valor não permitido. Além disso, a função deverá
plotar um gráfico da função no intervalo entre ±10, destacando neste gráfico os pontos
relacionados ao valor de passado para a função.
𝑓(𝑥) = 1 / √2 − 𝑥

```python
import math
import matplotlib.pyplot as plt
import numpy as np


def f(x):
    if x >= 2:
        raise Exception("Valor inválido.")
    else:
        return 1 / math.sqrt(2 - x)


vetX = np.linspace(-10, 1, 1000)
vetY = [f(i) for i in vetX]

i = float(input("Digite um valor -> "))


plt.plot(vetX, vetY)
plt.scatter(i, f(i), color="red")
plt.show()
```

# Questão 2
Utilizando a linguagem Python, e somente as bibliotecas Matplotlib e Numpy, crie uma
função para plotar uma raquete de tênis de mesa. No seu gráfico a raquete deve ser plotada
de forma que seja possível observar tanto o cabo quanto a área de rebatimento.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches


def plot_table_tennis_racket():
    fig, ax = plt.subplots()

    # Cabo da raquete
    handle = patches.Rectangle((1.77, -1), 0.5, 2, facecolor='black')
    ax.add_patch(handle)

    # Área de rebatimento (círculo)
    rebound_area = patches.Circle(
        (2, 2.5), 2, facecolor='red', edgecolor='black', linewidth=1)
    ax.add_patch(rebound_area)

    # Configurações do gráfico
    ax.set_xlim(-1, 5)
    ax.set_ylim(-1, 5)
    ax.set_aspect('equal', adjustable='box')
    plt.show()


# Chamar a função
plot_table_tennis_racket()
```

# Questão 3
Utilizando a linguagem Python, nenhuma biblioteca extra e List Comprehension crie uma função que devolva o resultado da fórmula a seguir para 𝑛 elementos.

Captura de tela 2023-04-24 205830.png

```python
def f(n):
    return 1 / (1 + (2 ** n))

def somatorio(n):
    soma = 0
    for i in range(1, n+1):
        soma += f(i)
    return soma

def resposta(n):
    return 4 * (1 - somatorio(n))

i = input("Digite um valor: ")

print(resposta(i))
```

# Questão 4

Usando a linguagem Python, sem utilizar as bibliotecas Sympy e Numpy, escreva uma função que receba um valor 𝑛 no domínio dos números reais e devolva o limite da função dada a seguir quando 𝑥 tende a 𝑛.

𝑓(𝑥) = 3𝑥2 − 1 / √𝑥

```python
def limite(x):
    n = 3 * (x ** 2) - 1
    d = x ** (1/2) + x

    if (x == 0):
        return

    return (n/d)


i = input("Digite um valor: ")
print(limite(i))
```
